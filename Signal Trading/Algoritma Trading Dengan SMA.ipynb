Deskripsi : Program ini menggunakan dua moving average untuk menentukan open posisi buy atau sell

```
# import library
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```
# load data
from google.colab import files
uploaded = files.upload()
```

```
#simpan data di variable
AAPL = pd.read_csv('AAPL.csv')
#tampilkan data
AAPL
```

```
#tampilkan data pada grafik
plt.figure(figsize=(12.5, 4.5))
plt.plot(AAPL['Adj Close'], label = 'AAPL')
plt.title('Apple adj. Close Price History')
plt.xlabel('Sep. 09, 2014 - Mar. 29, 2018')
plt.ylabel('Adj. Close Price USD ($)')
plt.legend(loc='upper left')
plt.show()
```

```
#Membuat garis simple moving average dalam 30 hari
SMA30 = pd.DataFrame()
SMA30['Adj Close'] = AAPL['Adj Close'].rolling(window= 30).mean()
SMA30
```

```
#Membuat garis simple moving average 100 
SMA100 = pd.DataFrame()
SMA100['Adj Close'] = AAPL['Adj Close'].rolling(window=100).mean()
SMA100
```

```
#tampilkan pada grafik
plt.figure(figsize=(12.5, 4.5))
plt.plot(AAPL['Adj Close'], label = 'AAPL')
plt.plot(SMA30['Adj Close'], label = 'SMA30')
plt.plot(SMA100['Adj Close'], label = 'SMA100')
plt.title('Apple adj. Close Price History')
plt.xlabel('Sep. 09, 2014 - Mar. 29, 2018')
plt.ylabel('Adj. Close Price USD ($)')
plt.legend(loc='upper left')
plt.show()
```

```
#Buat data frame baru untuk menyimpan semua data
data = pd.DataFrame()
data['AAPL'] = AAPL['Adj Close']
data['SMA30'] = SMA30['Adj Close']
data['SMA100'] = SMA100['Adj Close']
data
```

```
#membuat fungsi signal untuk mengindikasikan open position buy atau sell 
def buySell(data):
  sigPriceBuy = []
  sigPriceSell = []
  flag = -1

  for i in range(len(data)):
    if data['SMA30'][i] > data['SMA100'][i]:
      if flag != 1:
        sigPriceBuy.append(data['AAPL'][i])
        sigPriceSell.append(np.nan)
        flag = 1
      else:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(np.nan)
    elif data['SMA30'][i] < data['SMA100'][i]:
      if flag != 0:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(data['AAPL'][i])
        flag = 0
      else:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(np.nan)
    else:
      sigPriceBuy.append(np.nan)
      sigPriceSell.append(np.nan)

  return (sigPriceBuy, sigPriceSell)
```

```
#simpan data buy dan sell dalam suatu variabel
buySell = buySell(data)
data['Buy_Signal_Price'] = buySell[0] 
data['Sell_Signal_Price'] = buySell[1]
```

```
#tampilkan data
data
```

```
#tampilkan pada grafik
plt.figure(figsize=(12.6, 4.6))
plt.plot(AAPL['Adj Close'], label = 'AAPL', alpha = 0.35)
plt.plot(data['SMA30'], label = 'SMA30', alpha = 0.35)
plt.plot(data['SMA100'], label = 'SMA100', alpha = 0.35)
plt.scatter(data.index, data['Buy_Signal_Price'], label = 'Buy', marker = '^', color='green')
plt.scatter(data.index, data['Sell_Signal_Price'], label='Sell', marker='v', color='red')
plt.title('Apple adj. Close Price History, Sinyal Buy Dan Sell ')
plt.xlabel('Sep. 09, 2014 - Mar. 29, 2018')
plt.ylabel('Adj. Close Price USD ($)')
plt.legend(loc='upper left')
plt.show()
```
